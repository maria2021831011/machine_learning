data visualization library

```python
#importing some libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#seaborn has some buildin datasets
#total bill vs tip dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
#seting a theme for the plots
sns.set_theme()
#visualize the data
sns.relplot(data=tips, x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#iris
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
```

```python
#import french from
from sklearn.datasets import load_iris
iris=load_iris()
iri=pd.DataFrame(iris.data,columns=iris.feature_names)
iri['gender']=iris.target
iri.head()
```

```python
print(iri)
```

```python
#distribution plot
sns.distplot(iri['sepal length (cm)'])
```

```python
#colleration matrix
#creating heat map
cor=iri.corr()
sns.heatmap(cor,cbar=True,annot=True,square=True,annot_kws={'size':8},fmt='.1f',cmap='Blues')
```
