```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
#import numpy data for our plots
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

ploting the data

```python
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x axis and y axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y,'r.')
plt.show()
```

```python
plt.plot(x,y,'g--')
plt.show()
```

```python
plt.plot(x,y,'yx')
plt.show()
```

```python
# 2 wave
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.show()
```

```python
#bar plot
f=plt.figure()
ax=f.add_axes([0,0,1,1])
lang=['c','c++','java','python','javascript']
students=[100,170,60,30,115]
ax.bar(lang,students)
ax.set_xlabel('courses')
ax.set_ylabel('students')
ax.set_title('bar plot')
plt.show()
```

```python
#pi chart
f1=plt.figure()

```

```python
#bar plot
f=plt.figure()
ax=f.add_axes([0,0,1,1])
lang=['c','c++','java','python','javascript']
students=[100,170,60,30,115]
ax.pie(students,labels=lang,autopct='%1.1f%%')
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
f2=plt.figure()
ax=f2.add_axes([0,0,1,1])
ax.scatter(x,y,c='r')
ax.scatter(x,z,c='g')
plt.show()
```

```python
#3d scatter graph
g=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```
